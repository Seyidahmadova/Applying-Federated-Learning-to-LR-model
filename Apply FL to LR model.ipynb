```python
import pandas as pd
import numpy as np 
from numpy import log,dot,exp,shape
import matplotlib.pyplot as plt
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix,classification_report
import seaborn as sns
```

```python
data = pd.read_csv('C:\\Users\\user\\Documents\\Practice\\heart.csv')
```

```python
def encoding(x,y):
    ordinal_encoder = OrdinalEncoder()
    ordinal_encoder.fit(x)
    x = ordinal_encoder.transform(x)
        
    scaler = MinMaxScaler()
    x = scaler.fit_transform(x)  

    label_encoder = LabelEncoder()
    label_encoder.fit(y)
    y = label_encoder.transform(y)

    return x,y
```

```python
class LogisticRegression1:
    def __init__(self, learning_rate=0.01, num_iterations=500):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.weights = 0
        self.bias = 0
        self.arr = []

    def fit(self, X, y):
        
        # Initialize weights and bias
        num_samples, num_features = X.shape
        self.weights = np.zeros(num_features)
        self.bias = 0
        

        # Gradient descent optimization
        for _ in range(self.num_iterations):
            linear_model = np.dot(X, self.weights) + self.bias
            y_pred = self._sigmoid(linear_model)

            # Update weights and bias
            dw = (1 / num_samples) * np.dot(X.T, (y_pred - y))
            db = (1 / num_samples) * np.sum(y_pred - y)

            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db
            self.arr.append(self.weights.copy())
        return self.arr

    def predict(self, X):
        linear_model = np.dot(X, self.weights) + self.bias
        y_pred = self._sigmoid(linear_model)
        y_pred_class = np.where(y_pred >= 0.5, 1, 0)
        return y_pred_class

    def _sigmoid(self, x):
        return 1 / (1 + np.exp(-x))
```

```python
x_1 = data.iloc[0:183 , :-1]
y_1 = data.iloc[0:183, -1]

x, y = encoding(x_1, y_1)

X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(x, y, test_size=0.2, random_state=42)

obj_1 = LogisticRegression1()
model_1= obj_1.fit(X_train_1,y_train_1)
weight_1 = np.array(obj_1.arr)
# arr = []

y_pred_1 = obj_1.predict(X_test_1)
y_train_1 = obj_1.predict(X_train_1)
f1_score_tr_1 = f1_score(y_train_1,y_train_1)
f1_score_te_1 = f1_score(y_test_1,y_pred_1)  
accuracy_1 = accuracy_score(y_test_1, y_pred_1)
precision_1 = precision_score(y_test_1, y_pred_1)
recall_1 = recall_score(y_test_1, y_pred_1)

print(f"Accuracy: {accuracy_1}")
print(f"Precision: {precision_1}")
print(f"Recall: {recall_1}")  
print('f1_score_tr - ', f1_score_tr_1) 
print('f1_score_te - ', f1_score_te_1)

# print(weight_1)
```
